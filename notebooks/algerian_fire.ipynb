## Algerian Forest Fire Datasets

Data Set Information:  

The dataset incoludes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.  

The period from June 2012 to September 2012. The dataset includes 11 attributes and 1 output attribute(class). The 244 instances have been classified into fire(138 classes) and not fire(106 classes) classes.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1) ##header = 1 to ignore the first line of csv file
dataset.head()
```

```python
dataset.info() ## all are objects
```

## Data Cleaning

```python
dataset.isnull().sum()
```

```python
##missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on the region from 122th index, we can make a new column based on the region 

1 : " Bejaia Region Dataset "  
2 : " Sidi-bel abbes Region Dataset  
Add a new column with region

```python
dataset.loc[:122,"Region"] = 0 ## till 122 add a new column 'region' with value as 0
dataset.loc[122:,"Region"] = 1 ## after 122 'region' with value 1
## adding a new column
```

```python
dataset.info()
```

```python
dataset[['Region']] = dataset[['Region']].astype(int)
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
## Remove the null values
dataset = dataset.dropna().reset_index(drop=True)
```

```python
dataset.isnull().sum()
```

```python
dataset.iloc[[122]]
```

```python
dataset = dataset.drop(122).reset_index(drop=True)
```

```python
dataset.iloc[[122]]
```

```python
dataset.columns ## classes column has extra spaces
```

```python
## fixing spaces in the column name
dataset.columns = dataset.columns.str.strip()
dataset.columns
```

```python
## Change the required columns to integer datatype

dataset[['day','month','year','Temperature','RH','Ws']] = dataset[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
dataset.info()
```

```python
## Changing the other columns to float datatype
##using list comprehension
objects = [features for features in dataset.columns if dataset[features].dtype=='O'] ## O means object
objects
```

```python
## now picking all except Classes
for i in objects:
    if i!= 'Classes':
        dataset[i]=dataset[i].astype(float)
##classes is still object
dataset.info()
```

```python
dataset.describe()
```

```python
## saving this cleaned dataset into another csv
dataset.to_csv('Algerian_forest_fires_dataset_CLEANED.csv',index=False)
```

## Exploratory Data Analysis

```python
df = dataset.drop(['day','month','year'],axis=1)
```

```python
df
```

```python
## Categories in classes


df['Classes'].value_counts() ## there are many different fire and not fire even though they should've been same, so to change that we will add .contains('not fire') in next part
```

```python
## Encoding of the categories in classes

df['Classes'] = np.where(df['Classes'].str.contains('not fire'),0,1)  ##np.where(df[column]=='col name',0(if true),1(if false;i.e.fire))
```

```python
## plot density plot for all features

plt.style.use('seaborn-v0_8-notebook')
df.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
## Percentage for pie chart
percentage = df['Classes'].value_counts(normalize=True) * 100  ##normalize=True gives percentage in decimal (*100 for percentage)
percentage
```

```python
## plotting the pie chart
classlabels = ['Fire','Not Fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie chart of Classes")
```
